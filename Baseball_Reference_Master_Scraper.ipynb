### Import Libraries

```python
import urllib
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests
```

### Get Orioles Home Game Links

```python
def scrape_boxcore(url):
    link = 'http://www.baseball-reference.com/'
    orioles_home_game_urls = []
    
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    team_schedule  = soup.find("table", {"id":"team_schedule"})
    orioles_home_game_urls = []
    for row in team_schedule.find_all('tr')[1:]:
        try:
            for cell in row.find_all('td')[3]:  
                link = 'http://www.baseball-reference.com' + cell.get("href")
                link.encode('utf8')
    #             print link
                if str(link).find('BAL') != -1:
                    orioles_home_game_urls.append(str(link))
                else:
                    pass
        except:
            pass
    return orioles_home_game_urls
   
list_of_urls = scrape_boxcore('http://www.baseball-reference.com/teams/BAL/2016-schedule-scores.shtml')
```

### Get away and home pitchers

```python
def get_away_pitcher(list_of_urls):
    records = []
    for url in list_of_urls:
        print url
        # Make a request
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')
        pitchers_and_hitters  = soup.find_all("div", {"class":"table_container"})
        
        # Where is the stat line
        away_pitchers = pitchers_and_hitters[2].find_all("tr", {"normal_text"})
        pitchers = away_pitchers[1].find_all('td')
        
        record =[]
        for pitcher in pitchers:
            pitcher_stat =  pitcher.get_text()
            pitcher_stat.encode('utf8')
            record.append(pitcher_stat)
        records.append(record)

    return records
```

```python
def get_home_pitcher(list_of_urls):
    records = []
    for url in list_of_urls:
        print url
        # Make a request
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')
        pitchers_and_hitters  = soup.find_all("div", {"class":"table_container"})
        
        # Where is the stat line
        away_pitchers = pitchers_and_hitters[3].find_all("tr", {"normal_text"})
        pitchers = away_pitchers[1].find_all('td')
        
        record =[]
        for pitcher in pitchers:
            pitcher_stat =  pitcher.get_text()
            pitcher_stat.encode('utf8')
            record.append(pitcher_stat)
        records.append(record)

    return records
```

### Get home and away pitchers

```python
away_records = get_away_pitcher(list_of_urls)
away_records
home_records = get_home_pitcher(list_of_urls)
home_records
```

### Concatenate home and away data frames

```python
df_home = pd.DataFrame(data=home_records)
df_away = pd.DataFrame(data=away_records)
baseball = pd.concat([df_home, df_away])
```

```python
baseball
```

### Convert data frame to CSV

```python
baseball.to_csv('pitchers.csv')
```
